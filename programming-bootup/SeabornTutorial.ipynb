### Preliminaries

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
````

````python
#seaborn settings
sns.set_palette("deep", desat=.6)
sns.set_context(rc={"figure.figsize": (8,4)})
````

### Load Data Sets

````python
titanic = pd.read_csv('snsdata/titanic.csv')
````

````python
server = pd.read_csv('snsdata/serverdata.csv')
````

````python
mlb = pd.read_csv('snsdata/mlbsalaries.csv')
````

### 1: Histograms

````python
plt.hist(titanic.Age.dropna(),bins=25)
````

````python
plt.hist(server.Time, 25, color=sns.desaturate("indianred",1))
````

````python
plt.hist(mlb.Salary.dropna(), 25)
````

### 2: Boxplot

Seaborn has changed its plotting interface since Alfred Essa made this tutorial.  In particular, the plot orientation argument ```vert = True|False``` has been replaced with ```orient='v'|'h'```

````python
sns.boxplot(titanic.Age.dropna(), orient='v')
````

````python
#sns.boxplot(titanic.Age, titanic.Sex, vert=False)
sns.boxplot(titanic.Age, titanic.Sex, orient='h')
````

````python
sns.boxplot(server.Time, orient='v')
````

````python
sns.boxplot(mlb.Salary, orient='v')
````

### 3: Kernel Density Estimate Plot

````python
sns.kdeplot(titanic.Age.dropna(), shade=True)
````

````python
 sns.kdeplot(server.Time.dropna(), shade=True)
````

````python
sns.kdeplot(mlb.Salary.dropna(), shade=True)
````

````python
sns.distplot(titanic.Age.dropna())
````

````python
 sns.distplot(server.Time.dropna())
````

````python
 sns.distplot(mlb.Salary.dropna())
````

### 4: Violin Plot

````python
sns.violinplot(titanic.Age.dropna(), orient='v', inner='quartile')
````

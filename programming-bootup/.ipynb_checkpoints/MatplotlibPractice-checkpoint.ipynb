Import the necessary modules

````python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
````

Generate a histogram

The argument ```histtype='stepfilled'``` means that the plotting routine only generates the exposed parts of the histogram's bar lines and fills them in

````python
x = np.random.normal(size=1000)
fig, ax = plt.subplots()
H = ax.hist(x, bins=50, alpha=0.5, histtype='stepfilled')
````

Generate a pie plot

When declaring the subplots, setting the two components of the ```figsize``` argument equal will make the subplots square

The ```explode``` argument tuple indicates how far to remove each segment of the pie from the other segments

````python
fracs = [30, 15, 45, 10]
colors = ['b','g','r','w']
fig, ax = plt.subplots(figsize=(6,6))
pie = ax.pie(fracs, colors=colors, explode=(0.1, 0.075, 0.05, 0.025), 
            shadow=True, labels=['A','B','C','D'])
````

Generate an error bar plot

The formatting argument ```fmt``` lets the user specify the style of the data points (. for small dots, o for big dots, x is self-explanatory) and the color of the error bars

````python
x = np.linspace(0,10,30)
dy = 0.1
y = np.random.normal(np.sin(x), dy)

fig, ax = plt.subplots()
plt.errorbar(x, y, dy, fmt='.b')
````

Fill the area between two data sets

````python
x = np.linspace(0,10,1000)
y1 = np.sin(x)
y2 = np.cos(x)
fig, ax = plt.subplots()
ax.fill_between(x, y1, y2, where=(y1 < y2), color='red')
ax.fill_between(x, y1, y2, where=(y1 > y2), color='blue')
````

Generate a scatter plot

For more color maps (```cmap``` argument), visit http://matplotlib.org/examples/color/colormaps_reference.html

````python
x = np.random.random(50)
y = np.random.random(50)
c = np.random.random(50) # color of points
s = 500 * np.random.random(5) # size of points

fig, ax = plt.subplots()
im = ax.scatter(x, y, c=c, s=s, cmap=plt.cm.jet)
# Add a color bar
fig.colorbar(im, ax=ax)
# Setting a color limit apparently isn't necessary here, 
# but it can't hurt to know how it's done
im.set_clim(0.0, 1.0)
````

Countour plot with both lines and color variation

````python
x = np.linspace(0,10,50)
y = np.linspace(0,20,60)
````

````python
# This is the equivalent of a MatLab dot-multiply (.*)
z = np.cos(y[:, np.newaxis]) * np.sin(x)
````

````python
fig, ax = plt.subplots()
# Filled contours
im = ax.contourf(x, y, z, 100)
# Contour lines
# In this case, black lines
im2 = ax.contour(x, y, z, colors='k')
fig.colorbar(im, ax=ax)
````

Practice with the ```imshow``` plot

````python
I = np.random.random((100, 100))
I += np.sin(np.linspace(0, np.pi, 100))
fig, ax = plt.subplots()
im = ax.imshow(I, cmap=plt.cm.jet)
fig.colorbar(im, ax=ax)
````

2D histograms with both hexagonal and square bins

````python
x, y = np.random.normal(size=(2, 10000))

fig, ax = plt.subplots()
im = ax.hexbin(x, y, gridsize=25)
fig.colorbar(im, ax=ax)

fig, ax = plt.subplots()
H = ax.hist2d(x, y, bins=25)
fig.colorbar(H[3], ax=ax)
````

Polar plots and practice actually making use of subplot capabilities

````python
#fig = plt.figure()
#ax = fig.add_subplot(1, 1, 1, projection='polar')
fig, ax = plt.subplots(1, 2, subplot_kw=dict(projection='polar'))

theta = np.linspace(0, 10 * np.pi, 1000)
r = np.linspace(0, 10, 1000)
r2 = np.sin(2.0*theta)

ax[0].plot(theta, r)
ax[1].plot(theta, r2)
````
